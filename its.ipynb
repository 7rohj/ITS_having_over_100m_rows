```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
import pprint
import json

def parse_re(input_re: str):
    try:
        return_dict = json.loads(input_re)
    except json.decoder.JSONDecodeError:
        return_dict = dict()
    finally:
        return return_dict

url = 'https://openapi.its.go.kr:9443/trafficInfo?apiKey=b5142b3182d74c56846b70a346c0ff9e&type=all&minX=126.800000&maxX=127.890000&minY=34.900000&maxY=35.100000&getType=json'

response = requests.get(url)
contents = response.text
json_ob = parse_re(contents)
```
